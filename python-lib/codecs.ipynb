# codecs编解码函数

```python
import codecs
help(codecs)
```

## 1.直接调用

```python
a = '你好'
b = codecs.encode(a,encoding='gbk')
b2 = codecs.encode(a)
print(b)
print(b2)
```

```python
d = codecs.decode(b,encoding='gbk')
print(d)
d2 = codecs.decode(b2)
print(d2)
```

## 2.构造编码器

```python
look1 = codecs.lookup('gbk')
look2 = codecs.lookup('utf-8')

s = '我是天才'
print(s,len(s))
```

```python
d1 = look1.decode(bytes(s,'utf-8'))
print(d1,type(d1[0]))
```

```python
d2 = look2.decode(bytes(s,'utf-8'))
print(d2,type(d2[0]))
```

```python
print(len(d2[0])) # 此长度非彼长度
```

### 3.文件编码

```python
fileName = 'codecs.txt'

s = u'这是个unicode字符串'
print(s)
```

```python
f = codecs.open(fileName,'w')
f.write(s)
f.flush()
f.close()
```

以utf-8读取出来

```python
with codecs.open(fileName,'r','utf-8') as f:
    print(f.read())
```

以gbk读取出来

```python
with codecs.open(fileName,'r','gbk') as f:
    print(f.read())
```
