# collections提供一组有用的数据结构

```python
import collections

help(collections)
```

## 1.namedtuple

```python
Point = collections.namedtuple('Point',['x','y'])
print(Point.__doc__)
```

```python
p = Point(1,2)
print(p)
x,y = p
print(x,y)
```

```python
# 访问p
print(p[0],p.y)
```

```python
# 转化为dict
d = p._asdict()
print(d['x'])
# 或则从dict转过来
p = Point(**d)
print(p)
```

```python
# 替换,不会改变原对象
p = p._replace(x=10)
print(p)
```

## 2.deque

```python
help(collections.deque)
```

###  2.1 初始化deque

```python
q1 = collections.deque()
print(q1)
q2 = collections.deque([1,2,3,4,5])
print(q2)
q3 = collections.deque([1,4,2,5],5) # 最大长度为5
print(q3)
```

### 2.2 迭代

```python
for i in q2:
    print(i)
```

### 2.3 push pop

```python
q3.append([5,6,7])
print(q3)
q3.append(0)
print(q3)
```

**从上面结果可以看出，若声明了最大长度，会从后往前移动**
